```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from pathlib import Path
```

```python
# Reading images
filepath = str(Path("../images/birds.jpg"))
gray_scale_img = cv2.imread(filepath, 0)
rgb_scale_img = cv2.imread(filepath)
```

```python
cv2.imshow("gray_scale_img", gray_scale_img)
cv2.imshow("rgb_scale_img", rgb_scale_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Save the image
cv2.imwrite("gray_scale_img.jpg", gray_scale_img)
```

```python
# Reading image with matplotlib
img1 = plt.imread(filepath)
plt.imshow(img1)
```

### Draw the shape

```python
font = cv2.FONT_ITALIC

rectangle_draw = cv2.rectangle(rgb_scale_img, (384, 0), (510, 128), (0, 120, 255), 3)
cv2.putText(rectangle_draw, "Rectangle", (384+3, 0+30), font, 1, (225, 0, 0), 2)
cv2.imshow("rectangle_draw", rectangle_draw)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
